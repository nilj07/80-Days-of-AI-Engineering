# **READING LINE BY LINE**

```python
with open("data.txt") as f:
    for line in f:
        print(line.strip())
```

# **CSV FILE HANDLING**

```python
import csv

with open("students.csv", "r") as f:
    reader = csv.reader(f)
    for row in reader:
        print(row)
```

# **JSON FILE HANDLING**

```python
import json

with open("config.json") as f:
    data = json.load(f)

print(data["lr"])
```
