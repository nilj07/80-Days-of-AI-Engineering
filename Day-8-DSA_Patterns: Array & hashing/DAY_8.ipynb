# **ARRAY PATTERNS**

**Pattern 1: Traversal**

```python
arr = [1,5,2,3,6,4]
for i in range(len(arr)):
    pass
```

**Pattern 2: Prefix Sum**

```python
arr = [1, 2, 3, 4]
prefix = [0]

for x in arr:
    prefix.append(prefix[-1] + x)
```

# **HASHING**

**COMMON HASHING PATTERNS**

**Frequency Count**

```python
nums = [1,1,2,2,3]
freq = {}

for n in nums:
    freq[n] = freq.get(n, 0) + 1
```

**Membership Check**

```python
seen = set()

for x in nums:
    if x in seen:
        print("Duplicate found")
    seen.add(x)
```
