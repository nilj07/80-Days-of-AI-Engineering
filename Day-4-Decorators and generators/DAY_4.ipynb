# **Decorators**

```python
def my_decorators(func):
  def wrapper():
    print('before')
    func()
    print('after')
  return wrapper

@my_decorators
def say_hello():
  print('hello engineers')



say_hello()
```

# **DECORATOR WITH ARGUMENTS**

```python
def my_decorators(func):
  def wrapper(*args, **kwargs):
    print('before')
    result = func(*args, **kwargs)
    print('after')
    return result
  return wrapper


@my_decorators
def hi(name):
  print(f'hello {name}')

hi('Nilesh')
```

# **REAL-WORLD DECORATOR EXAMPLES**

```python
import time

def timer(func):
  def wrapper(*args, **kwargs):
    start = time.time()
    result = func(*args, **kwargs)
    end = time.time()
    print(end - start)
    return result
  return wrapper


@timer
def sleep(n):
  time.sleep(n)

sleep(2)
```

# **SWITCHING GEARS â†’ GENERATORS**

```python
def get_numbers():
    yield 1
    yield 2
    yield 3
```

```python
def read_file(path):
    with open(path) as f:
        for line in f:
            yield line
```
