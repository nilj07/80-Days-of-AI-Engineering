# **Local scope (inside function)**

```python
def func():
  x = 10
  print(x)

func()

# print(x) ---> gives error cause we access it outside the function
```

```python
print(x)
```

# **Enclosing scope**

```python
def outer():
    x = 20
    def inner():
        print(x)
    inner()

outer()
```

# **Global scope**

```python
x = 25

def func():
    print(x)

func()
```

**Modifying global**

```python
x = 10

def modify():
  global x
  x = 20
  print(x)

modify()

print(x)
```

# **Interview trap**

```python
x = 10

def outer():
    x = 20
    def inner():
        print(x)
    inner()

outer()
```

# **nonlocal KEYWORD**

Used to modify enclosing scope variable

```python
def outer():
  count = 0
  def inner():
    nonlocal count
    count = count + 1
    print(count)
  inner()

outer()
```

# **CLOSURES**

```python
def counter():
  count = 0
  def increment():
      nonlocal count
      count += 1
      return count
  return increment

c = counter()
print(c())
print(c())
print(c())
```

# **REAL-WORLD USE OF CLOSURES**

```python
def logger(level):
    def log(message):
        print(f"[{level}] {message}")
    return log

error_log = logger("ERROR")
error_log("File missing")
```

```python
def emp(name):
  def salary(salary):
    print(f'name: {name} \n salary: {salary}')
  return salary


employee = emp('Nilesh')
employee(126000)

empoyee = emp('XYZ')
employee(156250)
```
