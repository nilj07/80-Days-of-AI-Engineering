# **id() PROOF (INTERVIEW FAVORITE)**

```python
a = [1,2,3,4]

b = a
```

```python
# proof that a and b both assign one object

print(id(a))
print(id(b))
```

```python
a = 10
b = 10

print(id(a))
print(id(b))
```

# **SHALLOW COPY vs DEEP COPY (CRITICAL)**

```python
a = [[10,20],[30,40]]
```

```python
import copy

b = copy.copy(a)

b
```

```python
# shallow copy

a[0][0] = 100

print(b)
```

```python
# deep copy - it create fully independent copy

c = copy.deepcopy(a)

print(c)
```
