# **MEAN, MEDIAN, MODE**

**Mean**

```python
salaries = [10,8,15,11,6,7,9,15,90]
```

```python
mean = sum(salaries)/len(salaries)
mean
```

**Median (Middle Value)**

```python
import numpy
median = numpy.median(salaries)
median
```

# **Python implimentaion (STD)**

```python
import numpy as np

data = [5, 6, 7, 8, 50]

np.mean(data)
np.median(data)
np.var(data)
np.std(data)
```
