# **NUMPY – WHY IT EXISTS**

```python
a = [1,2,3]
b = [4,5,6]

a + b  # element wise adding not possible
```

**Solution**

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4,5,6])

a + b
```

# **NUMPY ARRAY BASICS**

```python
arr = np.array([1, 2, 3])
print(arr.shape)
print(arr.dtype)
```

# **INDEXING & SLICING**

```python
arr = np.array([10, 20, 30, 40])

arr[1]      # 20
arr[1:3]    # [20 30]
```

# **COMMON NUMPY FUNCTIONS**

```python
np.mean(arr)
np.sum(arr)
np.max(arr)
np.min(arr)
```
