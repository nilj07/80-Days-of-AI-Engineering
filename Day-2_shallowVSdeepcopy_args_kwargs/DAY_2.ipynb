# **SHALLOW vs DEEP COPY**

**Shallow Copy**

```python
import copy

a = [[5,10],[15,20]]

b = copy.copy(a)

print(id(a), id(b))     # both id are different
```

```python
print(id(a[0]))
print(id(b[0]))   # both values id are same
```

**Deep Copy**

```python
c = copy.deepcopy(a)

print(id(a[0]))
print(id(c[0]))    # both id are different
```

# ***args â€” VARIABLE POSITIONAL ARGUMENTS**

```python
def show(*args):
  print(args)

show(2,3,5,6,9,8,7)
```

**âœ… Useful Example**

```python
def add_nums(*args):
  total = 0
  for num in args:
    total = num + total
  return total

add_nums(5,2,4)
```

ðŸ”„ **UNPACKING WITH** - * opens the list and passes values one by one.

```python
numbers = (2,1,3,4)

add_nums(*numbers)
```

# ***kwargs â€” VARIABLE KEYWORD ARGUMENTS**

```python
def show_names(**kwargs):
  print(kwargs)

show_names(name="Nilesh", age=22, role="ML Engineer")
```
